**(Plot) Total number of responses**

```python
df['Response'].hist()
```

**(Plot) Rate of the Response types by each Sales Channel**

```python
plt.figure(figsize=(8,4))
sns.countplot(x='Sales Channel', hue='Response', data=df)
plt.ylabel('number of responses')
plt.title('Responses by Sales Channel')
plt.show()
```

**(Boxplot) Total Claim Amount for each Response type**

```python
sns.boxplot(x='Total Claim Amount', y='Response', data=df)
```

**(Distribution plot) Total Claim Amount for each Response type**

```python
sns.histplot(data=df, x='Total Claim Amount', hue='Response', multiple='stack')
```

**(Distribution plot) Income for each Response type**

```python
sns.histplot(data=df, x='Income', hue='Response', multiple='stack')
```

**(Scatterplot) for Total Claim Amount and Income**

```python
plt.scatter(x=df['Total Claim Amount'], y=df['Income'])
plt.show()
```
